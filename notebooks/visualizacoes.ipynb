```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("../data/raw/avaliacao_mensal_funcionarios.csv", decimal=",")
file.columns = ["cargo","setor","salarioBruto","tipoContratacao","cargaHoraria","formacao","idade","nps"]
```

```python
file.head()
```

```python
file.setor = file.setor.replace({"Engenheiro de Software":"Engenharia de Software"})
```

```python
mediaPorSetor = file.groupby("setor").mean(numeric_only=True).nps

plt.bar(mediaPorSetor.index, mediaPorSetor.values)
plt.ylabel("NPS Médio por setores")
plt.xticks(rotation = 30)
plt.yticks(np.arange(0,11,1))
plt.title("NPS médio mensal por setor");
```

```python
mediaPorContratacao = file.groupby("tipoContratacao").mean(numeric_only=True).nps

plt.bar(mediaPorContratacao.index, mediaPorContratacao.values)
plt.ylabel("NPS Médio por tipo de contratação")
plt.xticks(rotation = 30)
plt.yticks(np.arange(0,11,1))
plt.title("NPS médio mensal por tipo de contratação");
```

```python
plt.hist(file["nps"])
plt.title("Distribuição do NPS mensal")
plt.xlabel("NPS");
```
